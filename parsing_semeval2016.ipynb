```python
import xml.etree.ElementTree as ET

# Path to the XML file
xml_file = 'ABSA16_Restaurants_Train_SB1_v2.xml'

# Parse the XML file
tree = ET.parse(xml_file)
root = tree.getroot()

```

```python
texts = []
targets = []
polarities = []
for sentence in root.findall(".//sentence"):
    # Get children of sentence
    text = sentence.find('text').text
    Opinions = sentence.find('Opinions')
    try:
        opinition = Opinions.findall('Opinion')
        for op in opinition:
            target = op.attrib['target']
            polarity = op.attrib['polarity']
            texts.append(text)
            targets.append(target)
            polarities.append(polarity)
    # Print the stacktrace if happen
    except Exception as e:
        print("Error in sentence: ", text)
        print(e)
print(texts)
print(targets)
print(polarities)
    

```

```python
counter = 0
to_remove = []
print(len(texts))
for text, target, polarity in zip(texts, targets, polarities):
    if target.strip() == 'NULL':
        to_remove.append(counter)
    counter += 1
print(len(texts))
```

```python
texts = [text for i, text in enumerate(texts) if i not in to_remove]
targets = [target for i, target in enumerate(targets) if i not in to_remove]
polarities = [polarity for i, polarity in enumerate(polarities) if i not in to_remove]
```

```python
polarity_mapping = {'negative': -1, 'positive': 1, 'neutral': 0}
polarities = [polarity_mapping[polarity] for polarity in polarities]
polarities
```
